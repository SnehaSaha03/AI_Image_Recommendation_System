```python
from PIL import Image

def preprocess_image(image_path):
    image = Image.open(image_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt")
    return inputs
```

```python
import torch

def get_image_embedding(image_inputs):
    with torch.no_grad():
        image_features = model.get_image_features(**image_inputs)
    # Normalize the feature vector
    image_features /= image_features.norm(dim=-1, keepdim=True)
    return image_features
```

```python
def match_captions(image_features, captions):
    # Process captions
    text_inputs = processor(text=captions, return_tensors="pt", padding=True)

    with torch.no_grad():
        text_features = model.get_text_features(**text_inputs)

    # Normalize text features
    text_features /= text_features.norm(dim=-1, keepdim=True)

    # Compute cosine similarity between image features and each caption feature
    similarity = (image_features @ text_features.T).squeeze(0)

    # Pair captions with similarity scores
    scored_captions = list(zip(captions, similarity.cpu().tolist()))

    # Sort captions by descending similarity score
    scored_captions.sort(key=lambda x: x[1], reverse=True)

    return scored_captions
```

```python
def image_captioning(image_path, captions, top_k=5):
    # Preprocess the image
    inputs = preprocess_image(image_path)

    # Generate image features
    image_features = generate_image_embeddings(inputs)

    # Match captions
    scored_captions = match_captions(image_features, captions)

    # Return top-k captions
    return scored_captions[:top_k]
```

```python
from torch.nn.functional import cosine_similarity

def match_captions(image_features, captions, processor, model):
    # Tokenize captions
    text_inputs = processor(text=captions, return_tensors="pt", padding=True, truncation=True)

    with torch.no_grad():
        text_features = model.get_text_features(**text_inputs)

    # Normalize features
    image_features = image_features / image_features.norm(dim=-1, keepdim=True)
    text_features = text_features / text_features.norm(dim=-1, keepdim=True)

    # Compute cosine similarity
    similarities = cosine_similarity(image_features, text_features)

    return similarities
```

```python
def recommend_captions(image_path, captions, model, processor, top_k=5):
    # Preprocess image
    inputs = preprocess_image(image_path, processor)

    # Generate image features
    image_features = generate_image_embeddings(model, inputs)

    # Generate text features
    text_inputs = processor(text=captions, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        text_features = model.get_text_features(**text_inputs)

    # Compute cosine similarities
    similarities = torch.matmul(image_features, text_features.T)

    # Get top-k most similar captions
    top_indices = similarities[0].topk(top_k).indices

    results = []
    for idx in top_indices:
        caption = captions[idx]
        score = similarities[0][idx].item()
        results.append((caption, score))

    return results
```

```python
from PIL import Image
import torch
from transformers import CLIPProcessor, CLIPModel
import torch.nn.functional as F

# Load pre-trained CLIP model and processor
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

def recommend_captions(image_path, captions, model, processor, top_k=5):
    # Load and preprocess the image
    image = Image.open(image_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt")

    # Generate image embeddings
    with torch.no_grad():
        image_features = model.get_image_features(**inputs)
    image_features /= image_features.norm(dim=-1, keepdim=True)

    # Process captions and get text embeddings
    text_inputs = processor(text=captions, return_tensors="pt", padding=True)
    with torch.no_grad():
        text_features = model.get_text_features(**text_inputs)
    text_features /= text_features.norm(dim=-1, keepdim=True)

    # Compute cosine similarity between image and text embeddings
    similarities = (image_features @ text_features.T).squeeze(0)

    # Get top-k captions
    top_indices = similarities.topk(top_k).indices.tolist()

    results = [(captions[i], similarities[i].item()) for i in top_indices]

    return results

# Example usage:
image_path = "/WhatsApp Image 2025-06-09 at 16.02.33_a38d9af9.jpg"  # Replace with your image path
captions = [
    "A beautiful sunset over the mountains",
    "A cat sitting on a sofa",
    "A delicious plate of food",
    "A person riding a bike",
    "An airplane flying in the sky"
]

results = recommend_captions(image_path, captions, model, processor, top_k=5)

for caption, score in results:
    print(f"Caption: {caption} | Similarity Score: {score:.4f}")
```
